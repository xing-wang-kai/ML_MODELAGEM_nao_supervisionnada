# MACHINE LEARNING

## MODELO DE APRENDIZAGEM NÃO SUPERVISIONADOS

Este modelo de apreendizagem de maquina não tem um padrão de teste e precisa de clustering para validar semelhanças entre dados e categorizar o mesmo.

```python
import pandas as pd
import numpy as np

SEED = 123
np.random.seed(SEED)

url = "https://raw.githubusercontent.com/alura-cursos/alura-clustering-validation/base-de-dados/CC%20GENERAL.csv"

data = pd.read_csv(url, sep=',')
data
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.drop(['CUST_ID', 'TENURE'], axis=1, inplace=True)
data.head()
```

```python
data.fillna(data.median(), inplace=True)
data.isna().sum()
```

```python
from sklearn.preprocessing import Normalizer

dataFrame = Normalizer().fit_transform(data.values)
dataFrame
```

```python
## O parametro N_INIT é quantas vezes devem retornar o mesmo valor para que a
## para que o resultado possa ser deterministico

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, n_init=10, max_iter=300)
y_predict = kmeans.fit_predict(dataFrame)
```

```python
# As silhouettes são valores menores > 1 e maiores que -1 e que defimem o score do
# do agrupamento do K medias. Nesse caso o agrupamento foi de 0,36 superior ao valor
# de 0, o que garante uma boa precisão para o algoritmo selhecionado.
# Quanto mais próximo de 1 estiver o valor de silhouette mais separados estão os clusters.

from sklearn import metrics

labels = kmeans.labels_
silhouettes  = metrics.silhouette_score(dataFrame, labels, metric='euclidean')
silhouettes
```

```python
# Este modelo calcula dados entre o centroid dos clustters e retorna dados
# quanto mais próxmo de 0 melhor o modelo avaliativo do clutters

dbs = metrics.davies_bouldin_score(dataFrame, labels)
dbs
```

```python
## RECEBE ONSIDERAÇÃO DA DISPERSÃO ENTRE CLUSTERS
# USA CALCULOS DE VÁRIANÇAS

calinsk = metrics.calinski_harabasz_score(dataFrame, labels)
calinsk
```

```python
centroid = kmeans.cluster_centers_
centroid
```

```python
for cols in range(len(centroid[0])):
  print(data.columns.values[cols])
  print(f"{centroid[:, cols].var():.4f}")
  print("*"*50)
```
